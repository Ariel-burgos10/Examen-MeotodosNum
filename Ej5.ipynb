```python
import numpy as np
from scipy.optimize import bisect, newton

# polinomio
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Derivada
def df(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625

# Método de Bisección

# Raíz
r1 = bisect(f, -4, -3)

# Raíz
r2 = bisect(f, 1, 2)


# Método de Newton

r3 = newton(f, x0=-0.8, fprime=df)

# resultados
print("Raíz 1 (Bisección):", r1)
print("Raíz 2 (Newton):   ", r3)
print("Raíz 3 (Bisección):", r2)
```

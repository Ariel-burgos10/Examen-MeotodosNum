```python
import numpy as np

def f(x):
    return x**3 + 2*x**2 + 3*x + 1

def secant_method(f, x0, x1, tol=1e-4, max_iter=100):

    x_prev = x0
    x_curr = x1

    f_prev = f(x_prev)
    f_curr = f(x_curr)

    iter_count = 0

    while abs(f_curr) > tol and iter_count < max_iter:

        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)

        x_prev = x_curr
        f_prev = f_curr

        x_curr = x_next
        f_curr = f(x_curr)

        iter_count += 1

    return x_curr, iter_count

raiz, iteraciones = secant_method(f, 2, 3)

print("Raíz:", raiz)
print("Iteraciones:", iteraciones)
print("Llamadas a f:", iteraciones + 2)
```
